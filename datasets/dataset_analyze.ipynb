##Merge

```python
merge_df = df_emo.merge(df_explor, on=['sp_id', 'rp_id', 'seeker_post', 'response_post'], indicator=True, how='left')
merge_df_2 = df_emo.merge(df_explor, on=['sp_id', 'rp_id', 'seeker_post', 'response_post'], indicator=True, how='right')
```

```python
merge_df[merge_df['_merge']!='right_only']
```

```python
merge_df = pd.concat([merge_df, merge_df_2], ignore_index=True).drop_duplicates().drop(['len_rationales', '_merge'], axis=1)
```

```python
merge_df
```

```python
merge_df_3 = df_interp.merge(merge_df, on=['sp_id', 'rp_id', 'seeker_post', 'response_post'], indicator=True, how='left')
merge_df_4 = df_interp.merge(merge_df, on=['sp_id', 'rp_id', 'seeker_post', 'response_post'], indicator=True, how='right')
```

```python
pd.concat([merge_df_3, merge_df_4], ignore_index=True).groupby(['_merge']).count()
```

```python
merge_df = pd.concat([merge_df_3, merge_df_4], ignore_index=True).drop_duplicates().drop(['_merge'], axis=1)
```

```python
merge_df
```

```python
merge_df.loc[merge_df['level_interpretations'].isnull(), 'level_interpretations'] = 0
merge_df.loc[merge_df['level_emotional_reactions'].isnull(), 'level_emotional_reactions'] = 0
merge_df.loc[merge_df['level_explorations'].isnull(), 'level_explorations'] = 0
```

```python
merge_df.to_csv('Epitome.csv')
```
