##empathy detection in conversation

```python
df = pd.DataFrame()
df['utterance'] = epitome_df['seeker_post']
df['utterance_idx'] = 1
df['empathy'] = epitome_df['empathy']
df['conv_id'] = epitome_df['Unnamed: 0']
```

```python
df2 = pd.DataFrame()
df2['utterance'] = epitome_df['response_post']
df2['utterance_idx'] = 2
df2['empathy'] = epitome_df['empathy']
df2['conv_id'] = epitome_df['Unnamed: 0']
```

```python
conv_df = pd.concat([df, df2], ignore_index=True)
```

```python
conv_df = conv_df.sort_values(['conv_id']).reset_index()
```

```python
conv_df
```

```python
df_train, df_test = train_test_split(conv_df.conv_id.unique(), test_size=0.1, random_state=42)
df_train, df_val = train_test_split(df_train, test_size=0.1, random_state=42)
```

```python
conv_df_train = conv_df[conv_df['conv_id'].isin(df_train)].copy()
conv_df_test = conv_df[conv_df['conv_id'].isin(df_test)].copy()
conv_df_val = conv_df[conv_df['conv_id'].isin(df_val)].copy()
```

```python
df = pd.DataFrame()
df['utterance'] = emp_conv_df['utterance']
df['utterance_idx'] = emp_conv_df['utterance_idx']
df['empathy'] = emp_conv_df['empathy']
df['conv_id'] = emp_conv_df['conv_id']
```

```python
_df_train, _df_test = train_test_split(df.conv_id.unique(), test_size=0.1, random_state=42)
_df_train, _df_val = train_test_split(_df_train, test_size=0.1, random_state=42)
```

```python
df_train = df[df['conv_id'].isin(_df_train)].copy()
df_test = df[df['conv_id'].isin(_df_test)].copy()
df_val = df[df['conv_id'].isin(_df_val)].copy()
```

```python
conv_df_train = pd.concat([conv_df_train, df_train], ignore_index=True)
conv_df_test = pd.concat([conv_df_test, df_test], ignore_index=True)
conv_df_val = pd.concat([conv_df_val, df_val], ignore_index=True)
```

```python
conv_df_train.to_csv("train.csv")
conv_df_test.to_csv("test.csv")
conv_df_val.to_csv("val.csv")
```
